```python
bra = df[df['country']=='Brazil']
bra['year'].unique()
```

```python
jovens = (df.where(df['age']=='15-24 years')
    .dropna()
    .set_index(['year','country']))
jovens
```

```python
%%timeit -n 10
for group, frame in df.groupby('country'):
    avg = np.average(frame['population'])
    print('Counties in state ' + group + ' have an average population of ' + str(avg))
```

```python
k = df.groupby('year').agg({'suicides/100k pop': np.average})
k
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

with plt.style.context('dark_background'): #deixa dark
   
    k.plot(figsize=(18,10))
    plt.style.use('ggplot') #n sei oq faz/ deixa quadriculado
    plt.title('Taxa de suicídio de Vários Países ao longo dos anos') #adicionando o título
    plt.xlabel('Anos')
    plt.ylabel('Suicídio/100 mil')
    #plt.rcParams['figure.figsize'] = (11,7)
    plt.grid()
    plt.savefig('TotalTaxaSuicídio.png')
    
```
